```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
train_datagen = ImageDataGenerator(rescale=1./255, width_shift_range=0.1, height_shift_range=0.1)


# Load training data
train_generator = train_datagen.flow_from_directory(r'C:\Users\USER\Desktop\DATA\train', target_size=(224,224), batch_size=1, class_mode='categorical')

# Load validation data
validation_generator = train_datagen.flow_from_directory(r'C:\Users\USER\Desktop\DATA\test', target_size=(224,224), batch_size=1, class_mode='categorical')
```

```python
pretrained_model = tf.keras.applications.MobileNetV2(
    input_shape = (224,224,3),
    include_top = False,
    weights = 'imagenet',
    pooling = 'avg'

)

pretrained_model.trainable = False
```

```python
pretrained_model.summary()
```

```python
my_inputs = pretrained_model.input
```

```python
x = tf.keras.layers.Dense(128, activation='relu')(pretrained_model.output)
x = tf.keras.layers.Dense(128, activation='relu')(x)

my_outputs = tf.keras.layers.Dense(2, activation='softmax')(x)

model = tf.keras.Model(inputs = my_inputs, outputs = my_outputs)

model.compile( loss = 'categorical_crossentropy', optimizer = 'adam', metrics= ['accuracy'] )
```

```python
history = model.fit(
    train_generator,
    validation_data = validation_generator,
    epochs = 10,
    callbacks = [
        tf.keras.callbacks.EarlyStopping(
        monitor = 'val_loss',
        patience=3,
        restore_best_weights = True
        )
    ]
)
```

```python
import cv2 

img = cv2.imread("image047.jpg")
img = cv2.resize(img, (224,224))
img = img.reshape(1,224,224,3)

y = model.predict(img)[0]

print(y)
```
